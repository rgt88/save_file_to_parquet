Encrypted Parquet

```python
from cryptography.fernet import Fernet
import os

# Generate encryption key
key = Fernet.generate_key()
cipher = Fernet(key)

# List of parquet files to encrypt
file_list = input("Masukkan nama file parquet yang ingin dibaca (pisahkan dengan koma): ").split(",")

for file_parquet in file_list:
    file_parquet = file_parquet.strip()  # Remove extra spaces
    if not file_parquet.endswith(".parquet"):
        file_parquet += ".parquet"  # Add .parquet extension if not present
    
    if os.path.exists(file_parquet):
        # Baca file parquet
        with open(file_parquet, "rb") as file:
            parquet_data = file.read()

        # Enkripsi data
        encrypted_data = cipher.encrypt(parquet_data)

        # Simpan data terenkripsi ke file
        encrypted_filename = f"{file_parquet}_encrypted.parquet"
        with open(encrypted_filename, "wb") as encrypted_file:
            encrypted_file.write(encrypted_data)
        
        # Simpan kunci enkripsi (kunci ini diperlukan untuk dekripsi)
        key_filename = f"{file_parquet}_encrypted.key"
        with open(key_filename, "wb") as key_file:
            key_file.write(key)
        
        print(f"File {file_parquet} terenkripsi sebagai {encrypted_filename} dan kunci disimpan sebagai {key_filename}")
    else:
        print(f"File {file_parquet} tidak ditemukan.")
```
